```python
#Your code below:
import pandas as pd
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess
from scipy.spatial.distance import cosine
import numpy as np
```

```python
# Load the dataset
rand = pd.read_csv('RAND_Database_of_Worldwide_Terrorism_Incidents.csv', encoding='ISO-8859-1')

# Preprocess the text data
rand['desc'] = rand['Description'].fillna('')  # Replace missing descriptions with an empty string
rand['Processed_Description'] = rand['desc'].apply(simple_preprocess)  # Tokenize and clean text

# Define seed words for each category to guide clustering
political_seed_words = ["government", "protest", "election", "policy", "authority"]
religious_seed_words = ["religion", "faith", "mosque", "church", "islamic", "christian"]
economic_seed_words = ["economy", "market", "trade", "finance", "inflation"]

# Add seed words as extra sentences to the Word2Vec training data to anchor category meanings
additional_sentences = [political_seed_words, religious_seed_words, economic_seed_words]
all_sentences = rand['Processed_Description'].tolist() + additional_sentences

# Train the Word2Vec model with seed words added
w2v = Word2Vec(sentences=all_sentences, vector_size=100, window=5, min_count=2, workers=4)
# vector_size=100: Each word is represented by 100 numbers (dimensions)
# window=5: Model looks at 5 words before and after each word to understand its context
# min_count=2: Ignores words that appear less than twice to avoid rare, uninformative words
# workers=4: Uses 4 CPU cores for faster training

# Get the word vectors from the model
vectors = w2v.wv

def calculate_similarity(words, seed_words):
    """
    Calculates the average cosine similarity between a list of words from an event description
    and a list of seed words representing a category.

    Parameters:
    - words (list of str): List of words (tokens) from an event description.
    - seed_words (list of str): List of seed words representing a specific category (e.g., political).

    Returns:
    - float: The average similarity score between the description words and seed words. 
             Higher values indicate stronger similarity to the category.
    """
    # Get vectors for seed words in the model's vocabulary
    seed_vectors = [vectors[word] for word in seed_words if word in vectors]
    if not seed_vectors:  # If no seed words are in the vocabulary, return 0
        return 0
    # Get vectors for description words in the model's vocabulary
    word_vectors = [vectors[word] for word in words if word in vectors]
    if not word_vectors:  # If no description words are in the vocabulary, return 0
        return 0
    # Calculate cosine similarity between each description word and each seed word
    similarities = [1 - cosine(word_vector, seed_vector) for word_vector in word_vectors for seed_vector in seed_vectors]
    # Return the average similarity score, or 0 if there are no valid similarities
    return np.mean(similarities) if similarities else 0

def categorize_description(words):
    """
    Categorizes a description based on its similarity to political, religious, and economic categories.
    Similarity is determined by calculating average cosine similarity to seed words for each category.

    Parameters:
    - words (list of str): List of words (tokens) from an event description.

    Returns:
    - str: The category label based on similarity, with possible values:
           "Political", "Religious", "Economic", "Combination", or "Uncategorized".
    """
    # Calculate similarity scores for each category
    political_score = calculate_similarity(words, political_seed_words)
    religious_score = calculate_similarity(words, religious_seed_words)
    economic_score = calculate_similarity(words, economic_seed_words)

    # Store scores in a dictionary for easier comparison
    scores = {"Political": political_score, "Religious": religious_score, "Economic": economic_score}
    max_category = max(scores, key=scores.get)  # Find the category with the highest similarity score

    # Sort scores to compare top two categories
    sorted_scores = sorted(scores.values(), reverse=True)
    # If the top two scores are close, label it as "Combination"
    if sorted_scores[0] - sorted_scores[1] < 0.1:
        return "Combination"
    # If no category has a significant score, label as "Uncategorized"
    elif sorted_scores[0] == 0:
        return "Uncategorized"
    else:
        return max_category

# Apply categorization to each incident description
rand['Category'] = rand['Processed_Description'].apply(categorize_description)

# Display categorized data to check if it works as expected
print(rand[['Description', 'Category']].head())
```

```python
# Dictionary to store words in each cluster
word_clusters = {"Political": [], "Religious": [], "Economic": []}

# Loop through each word in the vocabulary
for word in vectors.index_to_key:
    # Calculate similarity of the word to each category
    political_score = calculate_similarity([word], political_seed_words)
    religious_score = calculate_similarity([word], religious_seed_words)
    economic_score = calculate_similarity([word], economic_seed_words)

    # Assign the word to the category with the highest similarity score
    scores = {"Political": political_score, "Religious": religious_score, "Economic": economic_score}
    best_category = max(scores, key=scores.get)
    word_clusters[best_category].append(word)

# Display sample words in each cluster to check clustering
for category, words in word_clusters.items():
    print(f"{category} Cluster: {words[:40]}")  # Display the first 40 words in each cluster
```
